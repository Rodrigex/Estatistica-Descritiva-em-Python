```python
def soma(a,b):
  return a+b
```

```python
soma(5,8)
```

```python
def subtrair(a,b):
  sub = a - b
  return sub
```

```python
subtrair(4,2)
```

```python
def nome_idade(nome, idade):
  resp = 'O nome é : '+nome+', e a idade é : '+str(idade)
  return resp
```

```python
nome_idade('Pedro', 70)
```

```python
 def subSoma(x,y):
   sub = subtrair(x, y)
   som = soma(x,y)

   return ('Subtração =' +str(sub)+', Soma =' +str(som))
```

```python
subSoma(6,7)
```
